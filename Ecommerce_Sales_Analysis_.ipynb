```python
import pandas as pd
import random


random.seed(42)
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}
df = pd.DataFrame(data)

# 1.
regional_sales = df.groupby('region')['sales_amount'].sum().reset_index()
regional_sales = regional_sales.sort_values(by='sales_amount', ascending=False)

# 2.
category_counts = df.groupby('category')['transaction_id'].count().reset_index()

# 3.
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()

print("Top Performing Region:\n", regional_sales.iloc[0])
```

```python
# 1.
region_cat_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

# 2.
salesperson_performance = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

# 3.
top_performer = salesperson_performance.sort_values(by='sum', ascending=False).iloc[0]

# 4.
cat_total = df.groupby('category')['sales_amount'].sum()
top_cat_name = cat_total.idxmax()
top_cat_value = cat_total.max()

print(f"Top Salesperson: {top_performer['salesperson']} with ${top_performer['sum']:.2f}")
print(f"Highest Revenue Category: {top_cat_name} (${top_cat_value:.2f})")
```

```python
# 1.
def sales_range(x):
    return x.max() - x.min()
regional_stats = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()

# 2.
summary_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("\nFinal Summary Report:")
print(summary_report)
```
