```python

import pandas as pd
from pandasql import sqldf


df = pd.read_csv('/content/employees_dataset_100_rows.csv')

sql_query = """
SELECT *
FROM df;
"""
employees = sqldf(sql_query, globals())
print(employees.to_markdown(index=False))
```

```python
# task 1

a="""
SELECT *
FROM df
WHERE department = 'Marketing';"""
print(sqldf(a, globals()))

b="""
SELECT name, department, salary
FROM df
WHERE salary > 90000;"""
print(sqldf(b, globals()))

c="""
SELECT *
FROM df
WHERE department = 'Sales'
OR department = 'Finance';"""
print(sqldf(c, globals()))
```

```python
# task 2
d="""
SELECT *
FROM df
ORDER BY salary DESC
LIMIT 5;"""
print(sqldf(d, globals()))

e="""SELECT *
FROM df
WHERE department = 'Engineering'
ORDER BY hire_date DESC
LIMIT 3;"""
print(sqldf(e, globals()))

f="""SELECT *
FROM df
WHERE salary > 70000
AND department != 'HR'
ORDER BY hire_date ASC
LIMIT 1;"""
print(sqldf(f, globals()))
```

```python
# task 3
g="""
SELECT
    name,
    employee_id,
    salary,
    hire_date
FROM df
WHERE salary > 60000;"""
print(sqldf(g, globals()))

h="""
SELECT
    name AS employee_name,
    salary AS annual_salary,
    salary / 12 AS monthly_salary,
    salary / (12 * 4) AS weekly_salary,
    salary / (12 * 4 * 30) AS daily_salary,
    salary / (12 * 4 * 30 * 60) AS hourly_salary,
    salary / (12 * 4 * 30 * 60 * 60) AS second_salary
    FROM df
    WHERE department = 'Finance'
    ORDER BY salary DESC;"""
print(sqldf(h, globals()))
```
